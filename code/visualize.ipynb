```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
odir_df = pd.read_csv("odir_labels.csv")
```

```python
x = list(odir_df.labels.unique())
y_before = [odir_df['labels'].value_counts()[label] for label in x]
y_after = [odir_df['labels'].value_counts()[label] for label in x]
```

```python
X_axis = np.arange(len(x))

plt.bar(X_axis-0.2, y_before, 0.4, label="Before GAN")
#plt.bar(X_axis+0.2, y_after, 0.4, label="After GAN")

# giving title to the plot
plt.title("Existence of Class Imbalance: ODIR")
    
# giving X and Y labels
plt.xticks(X_axis, x) 
plt.xlabel("Ocular Disease Label")
plt.ylabel("Number of Images")
    
# visualizing the plot
plt.legend()
plt.show()
```
