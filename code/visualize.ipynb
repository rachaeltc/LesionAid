```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
odir_df = pd.read_csv("../data/odir_labels.csv")
bal_odir_df = pd.read_csv("../data/balanced_odir_labels.csv")
```

```python
odir_df.columns
```

```python
x = list(odir_df.labels.unique())
y_before = [odir_df['labels'].value_counts()[label] for label in x]
y_after = [bal_odir_df['labels'].value_counts()[label] for label in x]
```

```python
for cla in zip(x, y_before):
    print(cla[0], cla[1])
```

```python
X_axis = np.arange(len(x))

plt.figure(figsize=(12,6))
plt.bar(X_axis-0.2, y_before, 0.4, label="Original ODIR-5K", color = 'slategray')
plt.bar(X_axis+0.2, y_after, 0.4, label="After Phase 1+2", color='mediumseagreen')

for i in range(len(x)):
    plt.text(i-0.2, y_before[i], y_before[i], ha = 'center')
    plt.text(i+0.2, y_after[i], y_after[i], ha = 'center')

# giving title to the plot
plt.title("Existence of Class Imbalance: ODIR")
    
# giving X and Y labels
plt.xticks(X_axis, x) 
plt.xlabel("Ocular Disease Label")
plt.ylabel("Number of Images")
    
# visualizing the plot
plt.legend()
plt.show()
```
