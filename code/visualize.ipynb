```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_df = pd.read_csv("../data/test_labels.csv")
train_df = pd.read_csv("../data/train_labels.csv")
train_rep_df = pd.read_csv("../data/train_rep_labels.csv")
#train_synth_df = pd.read_csv("../data/train_synth_labels.csv")
```

```python
x = list(test_df.labels.unique())
y_before = [train_df['labels'].value_counts()[label] for label in x]
y_rep_after = [train_rep_df['labels'].value_counts()[label] for label in x]
y_synth_after = [train_rep_df['labels'].value_counts()[label] for label in x]
```

```python
for cla in zip(x, y_before):
    print(cla[0], cla[1])
```

```python
X_axis = np.arange(len(x))

#plt.figure(figsize=(12,6))
plt.bar(X_axis-0.2, y_before, 0.4, label="Original ODIR-5K", color = 'slategray')
plt.bar(X_axis, y_synth_after, 0.4, label="Balanced with Synthesis", color='mediumseagreen')
plt.bar(X_axis+0.2, y_rep_after, 0.4, label="Balanced with Repetition", color='mediumseagreen')

for i in range(len(x)):
    plt.text(i-0.2, y_before[i], y_before[i], ha = 'center')
    plt.text(i, y_rep_after[i], y_rep_after[i], ha = 'center')
    plt.text(i+0.2, y_rep_after[i], y_rep_after[i], ha = 'center')

# giving title to the plot
plt.title("Existence of Class Imbalance: ODIR")
    
# giving X and Y labels
plt.xticks(X_axis, x) 
plt.xlabel("Ocular Disease Label")
plt.ylabel("Number of Samples")
    
# visualizing the plot
plt.legend()
plt.show()
```
